```python
import sqlite3
```

```python
conn = sqlite3.connect('longchau_db.sqlite')
cursor = conn.cursor()
```

```python
def run(sql):
    cursor.execute(sql)
    rows = cursor.fetchall()
    for r in rows:
        print(r)
```

```python
run (
"""
SELECT COUNT(*)
FROM products

"""
)
```

```python
run (
"""
SELECT product_url , COUNT(*)
FROM products
GROUP BY product_url 
HAVING COUNT(*) > 1 
"""
)
```

```python
run (
"""
SELECT *
FROM products
WHERE discount_text != '' AND price > strike_price 
"""
)
# có giảm giá nhưng strike_price bằng 0
```

```python
run (
"""
SELECT DISTINCT unit
FROM products;
"""
)
```

```python
run (
"""
SELECT product_name, price, strike_price,
       (strike_price - price) AS discount
FROM products
WHERE strike_price > 0 AND price > 0
ORDER BY discount DESC

"""
)
```

```python
run (
"""
SELECT  product_name, price, origin_country
FROM products
ORDER BY price DESC
LIMIT 2
"""
)
```

```python
run (
"""
SELECT unit, COUNT(*)
FROM products
GROUP BY unit
"""
)
```

```python
run (
"""
SELECT *
FROM products
WHERE product_name LIKE '%Vitamin_C%';
"""
)   
```

```python
run(
"""
SELECT *
FROM products
WHERE price BETWEEN 100000 AND 200000
"""
)
```

```python
run(
"""
SELECT product_name, price
FROM products
ORDER BY price ASC;
""")
```

```python
run(
"""
SELECT product_name, price, strike_price,
       100.0 * (strike_price - price) / strike_price AS percent_discount
FROM products
WHERE strike_price > price AND strike_price > 0
ORDER BY percent_discount DESC
LIMIT 5;
""")
```

```python
run(
"""
    SELECT 
        CASE
            WHEN price < 50000 THEN '<50k' 
            WHEN price between 50000 AND 100000 THEN '50k-100k'
            ELSE '>100k'
        END AS group_price , COUNT(*)
    FROM products
    GROUP BY group_price
"""
)
```

```python
run ("""
    SELECT *
    FROM products
    WHERE product_url IS NULL or product_url == ''
"""
)
```

```python
run (
"""
    SELECT product_name, brand , price
    FROM products 
    WHERE origin_country LIKE '%Việt_Nam%'
    ORDER BY price DESC
"""

)
```

```python
run (
"""
SELECT 
    CASE
        WHEN expiry LIKE '%12%' THEN '1 năm'
        WHEN expiry LIKE '%24%' THEN '2 năm'
        WHEN expiry LIKE '%36%' THEN '3 năm'
        ELSE '5 năm'
    END AS han_su_dung,
    COUNT(*)
FROM products
GROUP BY han_su_dung;

 """

)
```
